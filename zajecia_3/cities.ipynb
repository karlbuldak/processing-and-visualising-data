```python
import pandas as pd
```

```python
cities = pd.read_csv("cities-in-poland.csv", sep=';')
cities
```

```python
cities.loc[:,['City','Province']].groupby(['Province']).count()
```

```python
cities.loc[:,['Province','Population']].groupby(['Province']).sum()
```

```python
cities.loc[:,['Province','Region','Population']].query("Province == 'opolskie'").groupby(['Region']).sum()
```

```python
cities.loc[:,['Province','Region','Population']].query("Province == 'opolskie'").groupby(['Region']).sum().sort_values(by='Population',ascending=False)
```
